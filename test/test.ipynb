```python
import torch
from torch.autograd import Variable
```

```python
num_in = 10
num_hidden = 50

x = torch.randn(num_in)
W = torch.randn(num_hidden, num_in)
b = torch.randn(num_hidden)

z = torch.matmul(W,x)+b
h = torch.max(z, torch.zeros(num_hidden))
```

```python
num_in = 10
num_hidden = 50

x = Variable(torch.randn(num_in), requires_grad=True)
W = Variable(torch.randn(num_hidden, num_in), requires_grad=True)
b = Variable(torch.randn(num_hidden), requires_grad=True)

z = torch.matmul(W,x)+b
h = torch.max(z, Variable(torch.zeros(num_hidden)))
loss = (torch.sum(h)-1).pow(2)

loss.backward()
print(x.grad.data)
print(W.grad.data)
print(b.grad.data)
```
